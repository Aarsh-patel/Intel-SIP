#After Dataset Pruning



```python
import os
count = []
name = []
c = 0
dir_path = r'/content/Drive/MyDrive/Intel SIP/E_Auto/Pruned'
for path in os.listdir(dir_path):
    name.append(path)
    for i in os.listdir(os.path.join(dir_path, path)):
      c+=1
    count.append(c)
    print(f"Total images of {path} are {c}")
    c = 0
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize = (10, 5))

# creating the bar plot
plt.bar(name, count, width = 0.4)

plt.xlabel("Car company")
plt.ylabel("No. of images")
plt.show()
```
