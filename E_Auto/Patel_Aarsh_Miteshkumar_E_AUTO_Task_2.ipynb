##Creating new images

```python
dir_path = r'/content/drive/MyDrive/Intel SIP/E_Auto/Pruned'
for path in os.listdir(dir_path):
    c=0
    for i in os.listdir(os.path.join(dir_path, path)):
      c+=1
      image = cv2.imread(os.path.join(os.path.join(dir_path, path),i), 0)
      half = cv2.resize(image, (400, 400))
      cv2.imwrite(os.path.join(os.path.join(dir_path, path),str(c),".jpeg"), half)
    print(f"Completed {path}")
```
