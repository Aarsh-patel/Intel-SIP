```python
def update_dataset():

  for symbol in ['BTC','ETH','LTC']:
    a = pd.read_csv(f"/content/drive/MyDrive/Intel SIP/Crypto_Guides/{symbol}.csv")
    dat = a['Time'][-1]
    obj=datetime.fromtimestamp(dat+3600)
    data = yf.download(f"{symbol}-USD", start=obj, interval='1h') # Getting hourly data staring this year
    d_data = dict()
    data = data.to_dict()
    d_data['Time'] = [str(i.strftime("%s")) for i in data['Open']]
    for j in data.keys():
      if j == "Adj Close" or j=='Volume':
        continue
      d_data[j] = []
      for i in data[j].keys():
        d = '{:.3f}'.format(round(float(data[j][i]),3)) # Rounding to 3 decimal places
        d_data[j].append(d)
    data = pd.DataFrame.from_dict(d_data)
    a.append(data)
    data.to_csv(f"/content/drive/MyDrive/Intel SIP/Crypto_Guides/{symbol}.csv",index=False)
  print("Dataset Updated successfully !!!")
  return
```

## After updating the dataset.

```python
update_dataset()
a = pd.read_csv(f"/content/drive/MyDrive/Intel SIP/Crypto_Guides/BTC.csv")
a.tail()
```

## Adding clean data for further use

```python
def get_clean_data(start_date):
  data = yf.download(f"BTC-USD", start=start_date, interval='1h')
  data.to_csv(f"BTC.csv")
  d_data = dict()
  data = data.to_dict()
  d_data['Time'] = [str(i.strftime("%s")) for i in data['Open']]
  for j in data.keys():
    if j == "Adj Close" or j=='Volume':
      continue
    d_data[j] = []
    for i in data[j].keys():
      d = '{:.3f}'.format(round(float(data[j][i]),3)) # Rounding to 3 decimal places
      d_data[j].append(d)
  data = pd.DataFrame.from_dict(d_data)
  print(data)
  return
```
