```python
import yfinance as yf
data_1 = dict()
for i in ['BTC','ETH','LTC']:
  data_1[i] = yf.download(f"{i}-USD", start="2023-01-01", interval='1h') # Getting hourly data staring this year
data = yf.download("BTC-USD", start="2023-01-01", interval='1h')
```

```python
data = data.to_dict()
```

```python
d_data = dict()
d_data['Time'] = [i.strftime("%s") for i in data['Open']]
for j in data.keys():
  if j == "Adj Close" or j=='Volume':
    continue
  d_data[j] = []
  for i in data[j].keys():
    d = '{:.3f}'.format(round(float(data[j][i]),3)) # Rounding to 3 decimal places
    d_data[j].append(d)
```

```python
import pandas as pd
data = pd.DataFrame.from_dict(d_data)
```
