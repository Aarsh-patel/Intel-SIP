```python
import pandas as pd
import matplotlib.pyplot as plt
import talib
```

```python
# Loding the dataset of all 3 coins
BTC = pd.read_csv(f"/content/drive/MyDrive/Intel SIP/Crypto_Guides/BTC.csv")
ETH = pd.read_csv(f"/content/drive/MyDrive/Intel SIP/Crypto_Guides/ETH.csv")
LTC = pd.read_csv(f"/content/drive/MyDrive/Intel SIP/Crypto_Guides/LTC.csv")
```

# Graphical represntation of trends and patterns

#BTC

```python
plt.figure(figsize=(15, 8))
plt.title('BTC Prices History')
plt.plot(BTC['Close'][-1000:])
out = talib.SMA(BTC['Close'],200)
if out[len(LTC)-1]<LTC['Close'][len(LTC)-1]:
  print("Long term Bullish Trend")
else:
  print("Long term Bearish Trend")
plt.plot(out[-1000:])
out = talib.SMA(BTC['Close'],100)
if out[len(LTC)-1]<LTC['Close'][len(LTC)-1]:
  print("Mid term Bullish Trend")
else:
  print("Mid term Bearish Trend")
plt.plot(out[-1000:])
out = talib.SMA(BTC['Close'],50)
if out[len(LTC)-1]<LTC['Close'][len(LTC)-1]:
  print("Short term Bullish Trend")
else:
  print("Short term Bearish Trend")
plt.plot(out[-1000:])
plt.legend(['Close', 'SMA200', 'SMA100', 'SMA50'], loc='lower right')
plt.xlabel('Epoch')
plt.ylabel('Prices ($)')
```
