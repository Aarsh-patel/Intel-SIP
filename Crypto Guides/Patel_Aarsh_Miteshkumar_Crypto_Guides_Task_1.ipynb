## https://www.coinbase.com/

Issues with scraping data from coinbase.

*   Coinbase may block IP address or account if they detect that we are scraping their site, as it is against their terms of service.
*   The data we scrape may not be accurate or up-to-date, as Coinbase may change their website layout or API without notice.


Issues with the API client library

* Login credentials required and may require authentication.
* Granularity of data is not present.
* Not regularly updated.

## https://finance.yahoo.com

Pros of using yahoo finance

* Easy to use [API library](https://pypi.org/project/yfinance/)
* Not rate limited.
* Granular data is available.
* Data is updated every 5 minutes.

```python
# Hourly data = '1h'
# Weekly data = '1wk'
# Monthly data = '1mo'
import yfinance as yf
data = dict()
for i in ['BTC','ETH','LTC']:
  data[i] = yf.download(f"{i}-USD", start="2023-01-01", interval='1h') # Getting hourly data
```

```python
print(data['BTC'].tail())
```
